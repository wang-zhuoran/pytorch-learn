```python
import torch

x_data = [1., 2., 3.]
y_data = [2., 4., 6.]

w = torch.Tensor([1.0])
w.requires_grad = True
# w = torch.Tensor([1.0], requires_grad = True)
```

```python
# 构建计算图
def forward(x):
    return x*w

def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y)**2
```

```python
print("prediction (before training)", 4, forward(4).item())
```

```python
learning_rate = 0.01

for epoch in range(100):
    for x, y in zip(x_data, y_data):
        l = loss(x,y)
        l.backward()
        gradient = w.grad #注意Tensor直接运算会构建计算图
        print('\tgradient: ', x, y, gradient.item())
        w.data = w.data - learning_rate*gradient

        w.grad.zero_()
    
    print("progress: ", epoch, l.item())

print("prediction (after training)", 4, forward(4).item())
```

```python
print(w.data, type(w.data))
print(w, type(w))
print(w.item(), type(w.item()))
print(w.grad.data, type(w.grad.data))
print(w.grad, type(w.grad))
```
